```python
from transformers import T5Tokenizer, T5ForConditionalGeneration, AdamW
import torch
tokenizer = T5Tokenizer.from_pretrained("t5-base")
model = T5ForConditionalGeneration.from_pretrained("t5-base")
```

```python
# Step: Split your data
from sklearn.model_selection import train_test_split

train_data, val_data = train_test_split(df, test_size=0.2)
val_data, test_data = train_test_split(val_data, test_size=0.5)

# Step 1: Preprocess your data
tokenizer = T5Tokenizer.from_pretrained('t5-base')

def preprocess_data(df):
    inputs = []
    outputs = []
    for i, row in df.iterrows():
        title = row['TranslatedRecipeName']
        ingredients = str(row['TranslatedIngredients'])
        instructions = str(row['TranslatedInstructions'])
        # input_text = f"title: {title}"
        input_text = f"recipe: {title}\ningredients: {ingredients}\n"
        output_text = instructions.strip() + " </s>"
        inputs.append(input_text)
        outputs.append(output_text)
    input_encodings = tokenizer(inputs, padding=True, truncation=True, max_length=512, return_tensors='pt')
    output_encodings = tokenizer(outputs, padding=True, truncation=True, max_length=1024, return_tensors='pt')
    # input_encodings = tokenizer(inputs, padding=True, return_tensors='pt')
    # output_encodings = tokenizer(outputs, padding=True, return_tensors='pt')
    return input_encodings, output_encodings
```

```python

train_inputs, train_outputs = preprocess_data(train_data)
val_inputs, val_outputs = preprocess_data(val_data)
test_inputs, test_outputs = preprocess_data(test_data)
```

```python
# Step 2: Fine-tune the T5 model

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

def compute_loss(model, inputs, outputs):
    labels = outputs['input_ids']
    labels[labels[:, :] == tokenizer.pad_token_id] = -100
    input_ids = inputs['input_ids'].to(device)
    attention_mask = inputs['attention_mask'].to(device)
    labels = labels.to(device)
    attention_mask_decoder = outputs['attention_mask'].to(device)
    loss = model(input_ids=input_ids, attention_mask=attention_mask, decoder_input_ids=labels, decoder_attention_mask=attention_mask_decoder, labels=labels).loss
    return loss
```

```python
from tqdm.auto import tqdm
model = T5ForConditionalGeneration.from_pretrained('t5-base')
model.to(device)

optimizer = AdamW(model.parameters(), lr=5e-5)
```

```python
num_epochs = 5
for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    for i in tqdm(range(len(train_inputs['input_ids']))):
        inputs = {k: v[i:i+1] for k, v in train_inputs.items()}
        outputs = {k: v[i:i+1] for k, v in train_outputs.items()}
        loss = compute_loss(model, inputs, outputs)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    print(f"Epoch {epoch+1} loss: {epoch_loss/len(train_inputs['input_ids']):.4f}")
```

```python
def generate_recipe(model, title, ingredients):
    model.eval()
    input_text = f"recipe: {title}\ningredients: {ingredients}\n"
    input_encoding = tokenizer(input_text, padding=True, truncation=True, max_length=128, return_tensors='pt')
    input_ids = input_encoding['input_ids'].to(device)
    attention_mask = input_encoding['attention_mask'].to(device)
    output_ids = model.generate(input_ids=input_ids, attention_mask=attention_mask, max_length=128, num_beams=4, early_stopping=True)
    output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return output_text

title = "Lemon Chicken"
ingredients = "4 boneless chicken breasts\n1/4 cup lemon juice\n1/4 cup olive oil\n1 tsp salt\n1 tsp black pepper\n1 tsp dried oregano\n1 tsp minced garlic"
recipe = generate_recipe(model, title, ingredients)
print(recipe)
```
